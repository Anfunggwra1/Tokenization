import csv

input_filename = 'bod.csv'
output_filename = 'bodo.csv'

def enclose_with_quotes(text):
    if text.startswith('"') and text.endswith('"'):
        return text
    return f'"{text}"'

with open(input_filename, 'r', newline='', encoding='utf-8') as input_file, \
     open(output_filename, 'w', newline='', encoding='utf-8') as output_file:

    csv_reader = csv.DictReader(input_file)
    fieldnames = csv_reader.fieldnames

    if 'Bodo' not in fieldnames:
        raise ValueError("Column 'Bodo' not found in the input CSV.")

    csv_writer = csv.DictWriter(output_file, fieldnames=fieldnames)
    csv_writer.writeheader()

    for row in csv_reader:
        bodo_sentence = row['Bodo']
        modified_sentence = enclose_with_quotes(bodo_sentence)
        row['Bodo'] = modified_sentence
        csv_writer.writerow(row)

```python
from collections import Counter

input_filename = 'tok.txt'

with open(input_filename, 'r', encoding='utf-8') as file:
    token_counts = Counter()
    total_tokens=0
    total_sentences=0
    
    for line in file:
        tokens = line.split()
        total_tokens += len(tokens)
        token_counts.update(tokens) 
        total_sentences += 1
        
average_length = total_tokens / total_sentences if total_sentences > 0 else 0
num_unique_tokens = len(token_counts)
print(f"Number of unique tokens: {num_unique_tokens}")
print(f"Total number of tokens: {total_tokens}")
print(f"Average sentence length (in tokens): {average_length:.2f}")
```
